```python
import speech_recognition as sr
from pydub import AudioSegment
from pydub.silence import split_on_silence

def listen_and_recognize():
    recognizer = sr.Recognizer()

    with sr.Microphone() as source:
        print("Listening...")
        audio = recognizer.listen(source)

        try:
            # Recognize speech using Google Web API (can use offline models too)
            print("Processing speech...")
            text = recognizer.recognize_google(audio)
            return text
        except sr.UnknownValueError:
            return "Sorry, I couldn't understand the audio."
        except sr.RequestError as e:
            return "Error with speech recognition service: {0}".format(e)
```

```python
import tkinter as tk
from tkinter import messagebox
import threading
import speech_recognition as sr
import pyttsx3

recognizer = sr.Recognizer()
engine = pyttsx3.init()  # Initialize text-to-speech engine

def update_text_in_main_thread(result):
    """Updates the label in the main thread with speech recognition result."""
    label.config(text=f"You said: {result}")
    button_listen.config(state=tk.NORMAL)  # Re-enable the listen button

def listen_and_recognize():
    """Runs the speech recognition in a separate thread."""
    global recognizer
    with sr.Microphone() as source:
        recognizer.adjust_for_ambient_noise(source)
        try:
            audio = recognizer.listen(source)
            try:
                result = recognizer.recognize_google(audio)
                root.after(0, update_text_in_main_thread, result)  # Update label in main thread
            except sr.UnknownValueError:
                root.after(0, update_text_in_main_thread, "Sorry, I could not understand the audio.")
            except sr.RequestError as e:
                root.after(0, update_text_in_main_thread, f"Could not request results; {e}")
        except Exception as e:
            root.after(0, update_text_in_main_thread, f"Error: {e}")

def start_listening():
    """Starts the background thread for listening."""
    button_listen.config(state=tk.DISABLED)  # Disable listen button while listening
    listen_thread = threading.Thread(target=listen_and_recognize)
    listen_thread.daemon = True  # Daemon thread will shut down with the main program
    listen_thread.start()

def convert_text_to_speech():
    """Convert the input text to speech."""
    text = text_input.get()
    if text.strip():
        engine.say(text)  # Convert text to speech
        engine.runAndWait()
        label.config(text="Text has been converted to speech!")
    else:
        label.config(text="Please enter some text.")

# Create window
root = tk.Tk()
root.title("ADAPTIVE AUDIO TRANSCRIPTION")

# Styling
root.geometry("500x500")
root.config(bg="#f4f4f4")

# Title
title = tk.Label(root, text="ADAPTIVE AUDIO TRANSCRIPTION", font=("Helvetica", 18, "bold"), bg="#f4f4f4", fg="#333")
title.pack(pady=20)

# Output label
label = tk.Label(root, text="Press the button and say something", font=("Helvetica", 14), bg="#f4f4f4", fg="#333")
label.pack(pady=20)

# Speak button
button_listen = tk.Button(root, text="Speak", command=start_listening, font=("Helvetica", 12), bg="#007BFF", fg="white", padx=20, pady=10, relief="flat")
button_listen.pack(pady=10)

# Text input field for text-to-speech
text_input = tk.Entry(root, font=("Helvetica", 14), width=30)
text_input.pack(pady=20)

# Convert button for text-to-speech
button_convert = tk.Button(root, text="Convert", command=convert_text_to_speech, font=("Helvetica", 12), bg="#28A745", fg="white", padx=20, pady=10, relief="flat")
button_convert.pack(pady=10)

# Run the application
root.mainloop()
```
